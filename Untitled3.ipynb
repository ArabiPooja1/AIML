```python
class pooja:
    def __init__(self):
        self.m=None
        self.b=None
    def fit(self,xtrain,ytrain):
        den=0
        num=0
        for i in range(xtrain.shape[0]):
            num=num+((xtrain[i]+xtrain.mean())*(ytrain[i]+ytest.mean()))
            den=den+((ytrain[i]+ytrain.mean())*(ytrain[i]+ytest.mean()))
        self.m=num/den
        self.b=xtrain.mean()+(self.m*ytrain.mean())
        print(self.m)
        print(self.b)
    def predict(self,xtest):
        print(xtest)
        return self.m*xtest+self.b
```
