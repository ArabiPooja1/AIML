```python
a=pd.read_csv("wine_data.csv",header=None,usecols=[2,3,4])
a.columns=['Alco',"vian","desi"]
a
```

```python
import seaborn as sns
```

```python
sns.kdeplot(a['Alco'])
```

```python
sns.kdeplot(a['vian'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
new=MinMaxScaler()
```

```python
new.fit(a)
x_label=new.transform(a)
```

```python
x_label=pd.DataFrame(x_label,columns=a.columns)
```

```python
import numpy as np

np.round(a.describe())
```

```python
np.round(x_label.describe())
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.kdeplot(x="Alco",data=a)
sns.kdeplot(x="vian",data=a)
plt.title("Bfter")
plt.subplot(1,2,2)
sns.kdeplot(x="Alco",data=x_label)
sns.kdeplot(x="vian",data=x_label)
plt.title("After")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(a['Alco'], ax=ax1)
sns.kdeplot(a['vian'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_label['Alco'], ax=ax2)
sns.kdeplot(x_label['vian'], ax=ax2)
plt.show()
```

# FINDING OUTLIER USING EMPIRE RULE

```python
a=pd.read_csv("placement.csv")
a
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(x="cgpa",data=a)
```

```python
sns.distplot(a["cgpa"])
```

```python
up=a['cgpa'].mean()+3*a['cgpa'].std()
up
```

```python
lp=a['cgpa'].mean()-3*a['cgpa'].std()
lp
```

```python
a[a["cgpa"]>up] 
```

```python
a[a['cgpa']<lp]
```

```python
a[ ( a["cgpa"] > 8.80 )| (a['cgpa'] < 5.11) ] 
```

# removeing the outliers1

```python
a1=a.copy()
```

```python
a1['cgpa']=np.where(a['cgpa']>up,
                               up,
                                  np.where(a['cgpa']>lp,
                                                        lp, 
                                                         a['cgpa']
                    ))
```
