```python
tr=ColumnTransformer(
    transformers=[
        ('t3',StandardScaler(),['weight'])
    ],remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
new=LinearRegression(fit_intercept=True,
    normalize='deprecated',
    copy_X=True,
    n_jobs=None,
    positive=False,)
```

```python
pred=new.predict(xtest)
```
