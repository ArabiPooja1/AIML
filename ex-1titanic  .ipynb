```python
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv("dataset/titanic.csv")
```

```python
a.head()
```

```python
a=a.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
a.isnull().sum()
```

```python
a=a.dropna(subset=['Survived','Fare','Embarked'])
```

```python
a.isnull().sum()
```

```python
x=a.iloc[:,1:]
```

```python
x
```

```python
y=a.iloc[:,:1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=1/3)
```

```python
xtrain
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
tr=ColumnTransformer(
    transformers=(
    ('tr1',OrdinalEncoder(categories=[[1,2,3]]),['Pclass']),
    ('tr2',OneHotEncoder(drop='first',sparse=False),['Sex','Embarked']),
    ('tr3',SimpleImputer(),['Age']))
)
```

```python
tr
```

```python
tr.fit_transform(xtrain)
```

```python
tr.fit_transform(xtest)
```

```python
b=tr.fit_transform(a)
b
```
