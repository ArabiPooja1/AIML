```python
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv("dataset/titanic.csv")
```

```python
a
```

```python
a.drop(['PassengerId','Name','Ticket','Fare','Cabin'],inplace=True,axis=1)
```

```python
a
```

```python
a=a.dropna(subset=['Embarked'])
```

```python
a.isnull().sum()
```

```python
a
```

```python
x=a.iloc[:,1:]
```

```python
x
```

```python
y=a.iloc[:,0:1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
tr=ColumnTransformer(
        transformers=[
            ('tr2',OneHotEncoder(drop='first',sparse=False),['Sex','Embarked']),
            ('tr3',SimpleImputer(),['Age'])
        ],remainder='passthrough')
```

```python
tr
```

```python
xtrain=tr.fit_transform(xtrain)
xtrain
```

```python
xtest=tr.fit_transform(xtest)
xtest
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
new=LogisticRegression( penalty='l2',
    dual=False,
    tol=0.0001,
    C=1.0,
    fit_intercept=True,
    intercept_scaling=1,
    class_weight=None,
    random_state=None,
    solver='lbfgs',
    max_iter=100,
    multi_class='auto',
    verbose=0,
    warm_start=False,
    n_jobs=None,
    l1_ratio=None,
)
```

```python
new.fit(xtrain,ytrain)
```

```python
pred=new.predict(xtest)
```

```python
pred
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

```python
accuracy_score(ytest,pred)
```

```python
confusion_matrix(ytest,pred)
```

```python
precision_score(ytest,pred,average='micro')
```

```python
recall_score(ytest,pred,average='micro')
```

```python
f1_score(ytest,pred,average='micro')
```
