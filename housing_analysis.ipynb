```python
import pandas as pd
import numpy as np
a=pd.read_csv("dataset/Housing.csv")
```

```python
x=a.drop(columns="furnishingstatus")
```

```python
a
```

```python
y=a.iloc[:,-1:]
```

```python
y
```

```python
x1=pd.get_dummies(x,drop_first=True)
```

```python
x1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
new=LabelEncoder()
```

```python
y=new.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x1,y,random_state=0,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
pre=new.predict(xtest)
```

```python
pre
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score
```

```python
recall_score(ytest,pre,average='micro')
```
