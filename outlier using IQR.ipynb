# Zscore technique

```python
a['Zscore']=(a['cgpa']-a['cgpa'].mean())/a['cgpa'].std()
```

```python
a[a['Zscore']>3]
```

```python
a[a['Zscore']<-3]
```

```python
a[(a['Zscore']>3)|(a['Zscore']<-3)]
```

```python
df=a.copy()
```

```python
ul=df['cgpa'].mean()+3*df['cgpa'].std()
ul
```

```python
ll=df['cgpa'].mean()-3*df['cgpa'].std()
ll
```

```python
df1=df.copy()
```

```python
df1['cgpa']=np.where(
    df1['cgpa']>ul,
    ul,
    np.where(
        df1['cgpa']<ll,
        ll,
        df1['cgpa']
            )
)
```

```python
df[(df['Zscore']>3)|(df['Zscore']<-3)]
```

# Quantile method

```python
a=pd.read_csv("placement.csv")
```

```python
a
```

```python
a.describe()
```

```python
sns.boxplot(a['cgpa'])
```

```python
sns.distplot(a['cgpa'])
```

```python
a['cgpa'].skew()
```

```python
Q1=a['cgpa'].quantile(0.25)
Q1
```

```python
Q3=a['cgpa'].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
ll=Q1-1.5*IQR
ll
```

```python
ul=Q3+1.5*IQR
ul
```

```python
a[ (a['cgpa']>ul) | (a['cgpa']<ll)]
```

```python
df=a.copy()
```

```python
df['cgpa']=np.where(
    df['cgpa']>ul,
    ul,
    np.where(
    df['cgpa']<ll,
    ll,
    df['cgpa']))
```

```python
df[ (df['cgpa']>ul) | (df['cgpa']<ll)]
```
