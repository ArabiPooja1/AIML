```python
class LR:
    def __init__(self):
        self.m=None
        self.b=None
    def fit(self,xtrain,ytrain):
        new=0
        den=0
        for i in range(xtrain.shape[0]):
            new=new+((xtrain[i]-xtrain.mean())*(ytrain[i]-ytrain.mean()))
            den=den+((ytrain[i]-ytrain.mean())*(ytrain[i]-ytrain.mean()))
        self.m=new/den
        self.b=xtrain.mean()-(self.m*ytrain.mean())
        print(self.m)
        print(self.b)
    def predict(self,xtest):
        print(xtest)
        return self.m*(xtest+self.b)
```
