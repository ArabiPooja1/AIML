# Graph show it is simple Regression so we can use simple linear regression

```python
x=a.iloc[:,0:1]
```

```python
x
```

```python
y=a.iloc[:,1:2]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
```

```python
new=LinearRegression(fit_intercept=True,normalize='deprecated',copy_X=True,n_jobs=None,positive=False,)
```

```python
new.fit(xtrain,ytrain)
```

```python
pred=new.predict(xtest)
```

```python
pred
```

```python
    from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2_score(ytest,pred)
```

```python
new.score(xtrain,ytrain)
```

```python
mean_squared_error(ytest,pred)
```

```python
mean_absolute_error(ytest,pred)
```
